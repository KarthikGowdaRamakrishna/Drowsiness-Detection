# Load model

**Loading Pre-trained 'ultralitices' model from torch hub where PyTorch models are hosted ***Link: https://pytorch.org/hub/ultralytics_yolov5/*****
**From which we are going to be using the baseline model to keep our model lightweight and quick**

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
cap = cv2.videoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    cv2.imshow('YOLO', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
