# 5. Train from scratch

```python
import uuid # uniquie id's
import os # to use operating system dependent functionality
import time # to use sleep b/w captures 
```

```python
IMAGES_PATH = os.path.join('data', 'images') #/data/images
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
        # Writes out image to file 
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

python train.py: This runs the train.py Python script, which is responsible for training the YOLOv5 model.

The required parameters for train.py execution are:

<br>--img 320: Sets the image size to 320x320 pixels for training.</br>
<br>--batch 16: Sets the batch size to 16. Batch size is the number of training examples used in one iteration.</br>
<br>--epochs 500: The number of training epochs. An epoch is one complete pass through the entire training dataset.</br>
<br>--data dataset.yml: Specifies the dataset configuration file. This file contains paths to training and validation datasets, number of classes, and class names.</br>
<br>--weights yolov5s.pt: Initializes training with pre-trained weights from the file yolov5s.pt. YOLOv5s is one of the smaller models of YOLOv5, making it faster but less accurate than larger models.(**no need to have this PyTorch weights pre-downloaded this will do it**)</br>
<br>--workers 2: Sets the number of worker threads for loading data to 2.</br>

# 6. load custom Model

**torch.hub.load(...):** This function is used to load a model from a GitHub repository. In this case, it's being used to load a model from the Ultralytics YOLOv5 repository.

The required parameters for **torch.hub.load(...)** function execution are:

<br>**--'ultralytics/yolov5':** This is the GitHub repository where the YOLOv5 model is hosted.</br>
<br>**--'custom':** This indicates that you are loading a custom model. When you specify 'custom', it means that you are not using one of the pre-trained standard models (like 'yolov5s', 'yolov5m', etc.) but a model that we trained.</br>
<br>**--path='runs/train/exp/weights/last.pt':** This is the path to our custom-trained model weights.The file ***last.pt*** is the PyTorch format file containing the weights of the model.</br>



```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='runs/train/exp/weights/last.pt', force_reload=True)
```

```python
img = os.path.join('data', 'images', 'awake.04de87c6-a697-11ee-ac33-b206fbf79ce9.jpg')
```
